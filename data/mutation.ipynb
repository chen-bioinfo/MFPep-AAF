```python
def simple_saturation_mutations(sequence, output_file="GSSSGRGDSPA_mutations.fasta"):
    """简化版的饱和突变生成器"""
    
    amino_acids = "ACDEFGHIKLMNPQRSTVWY"
    
    with open(output_file, 'w') as f:
        count = 0
        for i in range(len(sequence)):
            for aa in amino_acids:
                # 构建突变序列
                mutated = sequence[:i] + aa + sequence[i+1:]
                # 写入FASTA格式
                f.write(f">mutation_GSSSGRGDSPA_{count}\n")
                f.write(f"{mutated}\n")
                count += 1
    
    print(f"为序列 '{sequence}' 生成了 {count} 个饱和突变")
    print(f"保存到: {output_file}")

# 使用
simple_saturation_mutations("GSSSGRGDSPA")
```
